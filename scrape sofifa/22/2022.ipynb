```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup

k = []
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36'
}

base_urls = ["https://sofifa.com/players?type=all&lg%5B%5D=13&lg%5B%5D=16&lg%5B%5D=19&lg%5B%5D=31&lg%5B%5D=53"]

# Loop through pages with an offset of 12000
for offset in range(0, 2881 , 60):
    url = f"https://sofifa.com/players?type=all&lg%5B0%5D=13&lg%5B1%5D=16&lg%5B2%5D=19&lg%5B3%5D=31&lg%5B4%5D=53&offset={offset}"
    resp = requests.get(url, headers=headers)
    soup = BeautifulSoup(resp.content, 'lxml')
    player_rows = soup.select('table.table.table-hover.persist-area tbody tr')

    for omk in player_rows:
        id_ = omk.select_one("img[id]")["id"]
        name = omk.select_one('td.col-name a:nth-child(1) div').get_text(strip=True)
        team = omk.select_one(".col-name > div > a").get_text(strip=True)
        position = omk.select_one('td.col-name a:nth-child(3) ').get_text(strip=True)
        age = omk.select_one(".col-ae").get_text(strip=True)
        overall = omk.select_one(".col-oa").get_text(strip=True)
        potential = omk.select_one(".col-pt").get_text(strip=True)
        wage = omk.select_one("td.col.col-wg").get_text(strip=True)

        # Extract the value from the specified HTML element
        value = omk.select_one('td.col.col-vl').get_text(strip=True)
        img_tag = omk.select_one('img[title]')
        nationality = img_tag['title']
            

        k.append([id_, name, team,nationality, position, age, overall, potential, value, wage])

        print(f"Id: {id_}, Name: {name}, Team: {team},Nationality: {nationality}, Position: {position}, Age: {age}, Overall: {overall}, Potential: {potential}, Value: {value}, Wage: {wage}")
```

```python
columns = ["Id", "Name", "Team","Nationality", "Position", "Age", "Overall", "Potential", "Value","Wage"]

# สร้างรายชื่อผู้เล่น


# สร้าง DataFrame
df = pd.DataFrame(k, columns=columns)

# แสดง DataFrame
```
